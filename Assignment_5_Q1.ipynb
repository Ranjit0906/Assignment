# Model validation technique

## Two Techniques: 1. Collinearity Check & 2. Residual Analysis

```python
# 1) Collinearity Problem Check
# Calculate VIF = 1/(1-Rsquare) for all independent variables
```

```python
rsq_age=smf.ols('age~km+hp+cc+doors+gears+tax+weight',data=toyota_df).fit().rsquared
vif_age=1/(1-rsq_age)

rsq_km=smf.ols('km~age+hp+cc+doors+gears+tax+weight',data=toyota_df).fit().rsquared
vif_km=1/(1-rsq_km)

rsq_hp=smf.ols('hp~age+km+cc+doors+gears+tax+weight',data=toyota_df).fit().rsquared
vif_hp=1/(1-rsq_hp)

rsq_cc=smf.ols('cc~age+km+hp+doors+gears+tax+weight',data=toyota_df).fit().rsquared
vif_cc=1/(1-rsq_cc)

rsq_doors=smf.ols('doors~age+km+hp+cc+gears+tax+weight',data=toyota_df).fit().rsquared
vif_doors=1/(1-rsq_doors)

rsq_gears=smf.ols('gears~age+km+hp+cc+doors+tax+weight',data=toyota_df).fit().rsquared
vif_gears=1/(1-rsq_gears)

rsq_tax=smf.ols('tax~age+km+hp+cc+doors+gears+weight',data=toyota_df).fit().rsquared
vif_tax=1/(1-rsq_tax)

rsq_weight=smf.ols('weight~age+km+hp+cc+doors+gears+tax',data=toyota_df).fit().rsquared
vif_weight=1/(1-rsq_weight)
```

# Model Deletion Diagnostics (checking/detecting Outliers or Influencers)

## Two Techniques : 1. Cook's Distance & 2. Leverage value

```python
# 1. Cook's Distance: If Cook's distance > 1, then it's an outlier
# Get influencers using cook's distance
```

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
c
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(toyota_df)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#index and value of influencer where c is more than .5
(np.argmax(c),np.max(c))
```

```python
from statsmodels.graphics.regressionplots import influence_plot
```

```python
# 2. Leverage Value using High Influence Points : Points beyond Leverage_cutoff value are influencers
fig,ax=plt.subplots(figsize=(20,20))
fig=influence_plot(model,ax = ax)
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n ; k = no.of features/columns & n = no. of datapoints
k=toyota_df.shape[1]
n=toyota_df.shape[0]
leverage_cutoff = (3*(k+1))/n
leverage_cutoff
```

# Improving model

```python
# Creating a copy of data so that original dataset is not affected
toyo_new=toyota_df.copy()
toyo_new
```

```python
# Discard the data points which are influencers and reassign the row number (reset_index(drop=True))
toyota_new=toyo_new.drop(toyo_new.index[[80]],axis=0).reset_index(drop=True)
toyota_new
```
